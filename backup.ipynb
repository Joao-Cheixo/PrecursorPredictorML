```python
def get_precursors_in_pathway(map_id, data):
    """
    Get all compounds in a pathway based on the map ID and the precursors.
    """
    compounds = set()
    pathway_data = data[data["pathway"] == map_id]
    for _, row in pathway_data.iterrows():
        precursors = row["precursors"].split(";")
        compounds.update(precursors)
    return compounds

map_id = "map00960"
pathway_data = data[data["pathway"] == map_id]

pathway_compounds = get_precursors_in_pathway(map_id,data)

print("Precursors in Pathway", map_id)
for compound in pathway_compounds:
    print(compound)
```

```python
#Metodo 1

def find_path_from_source_to_target(graph, source_compound, target_compound):
    """
    Find a path from the source compound to the target compound in the graph.
    """
    paths = nx.shortest_path(graph, source_compound, target_compound)
    
    return paths


source_compound = "C00047"
target_compound = "C12450"

path = find_path_from_source_to_target(graph, source_compound, target_compound)

if path is not None:
    print("Path found:")
    print(" <-> ".join(path))
else:
    print("No path found.")
```

```python
#Método 2 

import re

def get_allcompounds_of_pathway(graph):
    
    pattern = r'^C\d{5}$'  
    allcomp = [node for node in graph if re.search(pattern, node)]
    return allcomp

nodes = get_allcompounds_of_pathway(graph)
print(nodes) 

value = 'C00047'

if value in nodes:
    print("Value found in the list")
else:
    print("Value not found in the list")
```

```python
#Método 3 

def check_path_from_compound_to_precursor(graph, map_id, data):
    """
    Check if there is a path from a compound to a precursor in the graph.
    """
    compounds = get_allcompounds_of_pathway(graph)
    precursors = get_precursors_in_pathway(map_id, data)
    
    path_found = []

    for compound in compounds:
        for precursor in precursors:
            try:
                path = find_path_from_source_to_target(graph, compound, precursor)
                if path:
                    path_str = " -> ".join(path)
                    print(f"Path found: {path_str}")
                    path_found.append(path)
            except nx.NodeNotFound:
                print(f"No path found from compound {compound} to precursor {precursor}.")
```

```python
def check_path_from_compound_to_precursor(graph, map_id, data):
    """
    Check if there is a path from a compound to a precursor in the graph.
    """
    compounds = get_allcompounds_of_pathway(graph)
    precursors = get_precursors_in_pathway(map_id, data)

    for compound in compounds:
        for precursor in precursors:
            try:
                path = find_path_from_source_to_target(graph, compound, precursor)
                if path:
                    return True
            except nx.NodeNotFound:
                pass

    return False
```

```python

def get_compounds_for_precursors(graph, map_id, data):
    """
    Get a dictionary mapping precursors to their associated compounds.
    """
    compounds = get_allcompounds_of_pathway(graph)
    precursors = get_precursors_in_pathway(map_id, data)
    
    precursor_compound_dict = {}

    for precursor in precursors:
        precursor_compound_dict[precursor] = []

    for compound in compounds:
        for precursor in precursors:
            try:
                path = find_path_from_source_to_target(graph, compound, precursor)
                if path:
                    precursor_compound_dict[precursor].append(compound)
            except nx.NodeNotFound:
                pass

    return precursor_compound_dict

compound_prec_dict = get_compounds_for_precursors(graph, map_id, data)

print(compound_prec_dict)

```

```python
def convert_to_smiles(compound_id):
   
    structure = get_compound_structure(compound_id)

    mol = Chem.MolFromMolBlock(structure)

    if mol is not None:
        smiles = Chem.MolToSmiles(mol)
        return smiles
    else:
        return None
    
compound_id = "C00092"  # glucose
smile = convert_to_smiles(compound_id)
print(smile)
    
    
```

```python
#keys têm de ser os ids do metabolito pros compostos serem identificáves 
def get_smiles_for_precursors(precursor_compound_dict):
    smiles_dict = {}

    for compounds in precursor_compound_dict.values():
        for compound in compounds:
            compound_structure = get_compound_structure(compound)
            if compound_structure is not None:
                mol = Chem.MolFromMolBlock(compound_structure)

                if mol is not None:
                    compound_smiles = Chem.MolToSmiles(mol)
                    smiles_dict[compound] = compound_smiles
                else:
                    smiles_dict[compound] = None

    return smiles_dict


compound_smiles_dict = get_smiles_for_precursors(compound_prec_dict)
print(compound_smiles_dict)

```

```python
import pickle

def get_precursors_for_compound(graph, map_ids, data):
    """
    Get a dictionary mapping compounds to their associated precursor for all map IDs.
    """
    compound_prec_dict = {}
    map_ids = data.iloc[:, 0].unique().tolist()

    for map_id in map_ids:
        graph = KeggNetworkGenerator.get_kegg_network(map_id)
        compounds = get_allcompounds_of_pathway(graph)

        precursors = get_precursors_in_pathway(map_id, data)

        for compound in compounds:
            if compound not in precursors:
                compound_prec_dict.setdefault(compound, [])

                for precursor in precursors:
                    try:
                        if precursor in graph and compound in graph:
                            path = find_path_from_source_to_target(graph, precursor, compound)
                            if path:
                                compound_prec_dict[compound] = list(set(compound_prec_dict[compound]) | set([precursor]))
                    except (nx.NodeNotFound, nx.NetworkXNoPath):
                        pass

    return compound_prec_dict


compound_prec_dict = get_precursors_for_compound(graph, map_ids, data)
print(compound_prec_dict)

pickle_file = "compound_prec_dict.pkl"

with open(pickle_file, "wb") as f:
    pickle.dump(compound_prec_dict, f)

print("Compound precursor dictionary saved as pickle:", pickle_file)

```

```python
def get_all_maps_for_compound(graph, map_ids, data):
    """
    Get a dictionary mapping compounds to their associated maps for all map IDs.
    """
    compound_map_dict = {}
    map_ids = data.iloc[:, 0].unique().tolist()

    for map_id in map_ids:
        graph = KeggNetworkGenerator.get_kegg_network(map_id)
        compounds = get_allcompounds_of_pathway(graph)

        precursors = get_precursors_in_pathway(map_id, data)

        for compound in compounds:
            if compound not in precursors:
                if compound not in compound_map_dict:
                    compound_map_dict[compound] = []
                if map_id not in compound_map_dict[compound]:
                    compound_map_dict[compound].append(map_id)

    return compound_map_dict

compound_map_dict2 = get_all_maps_for_compound(graph, map_ids, data)
print(compound_map_dict2)
```
